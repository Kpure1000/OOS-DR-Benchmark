```python
from glob import glob
import os
import json

img_path = 'imgs/subplots/runtime-prop-new'

dist_path = os.path.join(img_path, 'dist')
prop_path = os.path.join(img_path, 'prop')
structure_path = os.path.join(img_path, 'structure')
runtime_prop_path = os.path.join(img_path, 'runtime-prop')
name_dict={
    'syn1':0,
    'syn2':1,
    'syn4':2,
}
runtime_name_dict={
    'sensor':       0,
    'letter':       1,
    'bison':        2,
    'fashionMNIST': 3,
    'merchant':     4,
    'raid':         5,
    'secom':        6,
    'coil20':       7,
    'cnae9':        8,
    'fmd':          9,
    'isomapFace':   10,
    'dendritic':    11,
    'headpose':     12,
}
dist_files={}
for f in glob(os.path.join(dist_path, '*.png')):
    baseName = os.path.basename(f)
    [dataset_name, metric_idx, _] = baseName.split('-')
    dataset_idx = name_dict[dataset_name]
    if dataset_idx not in dist_files:
        dist_files[dataset_idx] = {}
    dist_files[dataset_idx][int(metric_idx)-1] = f
print(json.dumps(dist_files, indent=2))

prop_files={}
for f in glob(os.path.join(prop_path, '*.png')):
    baseName = os.path.basename(f)
    [dataset_name, metric_idx, _] = baseName.split('-')
    dataset_idx = name_dict[dataset_name]
    if dataset_idx not in prop_files:
        prop_files[dataset_idx] = {}
    prop_files[dataset_idx][int(metric_idx)-1] = f
print(json.dumps(prop_files, indent=2))

structure_files={}
for f in glob(os.path.join(structure_path, '*.png')):
    baseName = os.path.basename(f)
    [metric_idx, _] = baseName.split('-')
    structure_files[int(metric_idx)-1] = f
print(json.dumps(structure_files, indent=2))

runtime_prop_files={}
for f in glob(os.path.join(runtime_prop_path, '*.png')):
    baseName = os.path.basename(f)
    [dataset_name, metric_idx, _] = baseName.split('-')
    dataset_idx = runtime_name_dict[dataset_name]
    if dataset_idx not in runtime_prop_files:
        runtime_prop_files[dataset_idx] = {}
    runtime_prop_files[dataset_idx][int(metric_idx)-1] = f
print(json.dumps(runtime_prop_files, indent=2))
```

```python
from pptx import Presentation
from pptx.enum.shapes import MSO_SHAPE_TYPE
from pptx.shapes.group import GroupShape
from pptx.text.text import TextFrame 
from pptx.shapes.picture import Picture
from pptx_replace import replace_shape_with_picture
from pptx.parts.image import Image, ImagePart
from typing import Optional

ppt = Presentation('doc/dist.pptx')
for slide_idx, slide in enumerate(ppt.slides):
    print(f'Slide {slide_idx + 1}')
    pic_count = 0
    for shape in slide.shapes:
        if shape.shape_type == MSO_SHAPE_TYPE.GROUP:
            shape: Optional[GroupShape]
            for ele_shape in shape.shapes:
                if ele_shape.shape_type == MSO_SHAPE_TYPE.PICTURE:
                    ele_shape: Optional[Picture]
                    replace_shape_with_picture(ele_shape, dist_files[slide_idx][pic_count])
                    print(f'  Replaced pic {pic_count + 1} as {os.path.basename(dist_files[slide_idx][pic_count])}')
                    pic_count += 1

ppt.save('doc/dist2.pptx')

```

```python
from pptx import Presentation
from pptx.enum.shapes import MSO_SHAPE_TYPE
from pptx.shapes.group import GroupShape
from pptx.shapes.picture import Picture
from pptx_replace import replace_shape_with_picture
from pptx.parts.image import Image, ImagePart
from typing import Optional

ppt = Presentation('doc/runtime-prop.pptx')
for slide_idx, slide in enumerate(ppt.slides):
    slide.notes_slide.notes_text_frame.text = list(runtime_name_dict.keys())[slide_idx]
    print(f'Slide {slide_idx + 1}')
    pic_count = 0
    for shape in slide.shapes:
        if shape.shape_type == MSO_SHAPE_TYPE.PICTURE:
            shape: Optional[Picture]
            if pic_count in runtime_prop_files[slide_idx]:
                replace_shape_with_picture(shape, runtime_prop_files[slide_idx][pic_count])
                print(f'  Replaced pic {pic_count + 1} as {os.path.basename(runtime_prop_files[slide_idx][pic_count])}')
            pic_count += 1

ppt.save('doc/runtime-prop2.pptx')

```
